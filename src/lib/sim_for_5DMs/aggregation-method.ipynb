ここで
$$
    \begin{equation}
        p = \frac{1}{\sum_{i\in N} \frac{{w^{\ast}_{ki}}^{\rm L} + {w^{\ast}_{ki}}^{\rm U}}{2}} = 1
    \end{equation}
$$ 
として、新たに $v_{ki}^{\rm L} = p{w^{\ast}_{ki}}^{\rm L}, v_{ki}^{\rm U} = p{w^{\ast}_{ki}}^{\rm U}$ が問題(9)の最適解になるかどうかを考える。

$p$ の定義より $p=1>0$ であるので(26)式、(27)式、(31)式の両辺を $p$ 倍することで
$$
    \begin{align}
        ~~ & ~~ v_{ki}^{\rm L}\leq a_{kij}v_{kj}^{\rm U} \\
        ~~ & ~~ a_{kij}v_{kj}^{\rm L}\leq v_{ki}^{\rm U} \\
        ~~ & ~~ v_{ki}^{\rm U}\geq v_{ki}^{\rm L}\geq \epsilon
    \end{align}
$$

また、
$$
    \begin{align}
        ~~ & ~~ \sum_{j\in N\setminus i} v_{kj}^{\rm U} + v_{ki}v^{\rm L} = p(\sum_{j\in N\setminus i} {w^{\ast}_{kj}}^{\rm U} + {w^{\ast}_{ki}}^{\rm L})\geq p \\
        ~~ & ~~ \sum_{j\in N\setminus i} v_{kj}^{\rm L} + v_{ki}v^{\rm U} = p(\sum_{j\in N\setminus i} {w^{\ast}_{kj}}^{\rm L} + {w^{\ast}_{ki}}^{\rm U})\leq p 
    \end{align}
$$

(32)式より
$$
    \begin{align}
        ~~ & ~~ \sum_{i\in N} \frac{v_{ki}^{\rm L} + v_{ki}^{\rm U}}{2} = 1
    \end{align}
$$

よって $v_{ki}^{\rm L}, v_{ki}^{\rm U}$ は問題(9)の全ての制約を満たす。  
ここで、この $v_{ki}^{\rm L}, v_{ki}^{\rm U}$ よりも問題(9)の目的関数値を小さくするような解 $v^{\ast}_{k} = [{v^{\ast}_{k}}^{\rm L}, {v^{\ast}_{k}}^{\rm U}]$ が存在することを仮定する。 
このとき、
$$
    \begin{align}
        ~~ & ~~ \sum_{i\in N} (v_{ki}^{\rm L} - v_{ki}^{\rm U}) > \sum_{i\in N} ({v^{\ast}_{ki}}^{\rm L} - {v^{\ast}_{ki}}^{\rm U})
    \end{align}
$$
を満たす。すなわち、問題(10)の最適解 $w_{k}^{\ast}$ が問題(9)の最適解にはならないことを仮定する。  
$v_{k}^{\ast}$ は問題(9)の制約条件を満たすので以下を満たす。
$$
    \begin{align}
        ~~ & ~~ {v^{\ast}_{ki}}^{\rm L}\leq a_{kij}{v^{\ast}_{kj}}^{\rm U} \\
        ~~ & ~~ a_{kij}{v^{\ast}_{kj}}^{\rm L}\leq {v^{\ast}_{ki}}^{\rm U} \\
        ~~ & ~~ \sum_{j\in N\setminus i} {v^{\ast}_{kj}}^{\rm U} + {v^{\ast}_{ki}}^{\rm L}\geq p \\
        ~~ & ~~ \sum_{j\in N\setminus i} {v^{\ast}_{kj}}^{\rm L} + {v^{\ast}_{ki}}^{\rm U}\leq p \\
        ~~ & ~~ \sum_{i\in N} \frac{{v^{\ast}_{ki}}^{\rm L} + {v^{\ast}_{ki}}^{\rm U}}{2} = 1 \\
        ~~ & ~~ {v^{\ast}_{ki}}^{\rm U}\geq {v^{\ast}_{ki}}^{\rm L}\geq \epsilon, i\in N
    \end{align}
$$

このとき、 $w_{ki}^{\rm L} = \frac{{v^{\ast}_{ki}}^{\rm L}}{p}, w_{ki}^{\rm U} = \frac{{v^{\ast}_{ki}}^{\rm U}}{p}$ が問題(10)の最適解になるかどうかを考える。

$p$ の定義より $p=1>0$ であるので(40)式、(41)式、(45)式の両辺を $\frac{1}{p}$ 倍することで
$$
    \begin{align}
        ~~ & ~~ w_{ki}^{\rm L}\leq a_{kij}w_{kj}^{\rm U} \\
        ~~ & ~~ a_{kij}w_{kj}^{\rm L}\leq w_{ki}^{\rm U} \\
        ~~ & ~~ w_{ki}^{\rm U}\geq w_{ki}^{\rm L}\geq \epsilon
    \end{align}
$$

また、
$$
    \begin{align}
        ~~ & ~~ \sum_{j\in N\setminus i} w_{kj}^{\rm U} + w_{ki}v^{\rm L} = \frac{1}{p}(\sum_{j\in N\setminus i} {v^{\ast}_{kj}}^{\rm U} + {v^{\ast}_{ki}}^{\rm L})\geq 1 \\
        ~~ & ~~ \sum_{j\in N\setminus i} w_{kj}^{\rm L} + w_{ki}v^{\rm U} = \frac{1}{p}(\sum_{j\in N\setminus i} {v^{\ast}_{kj}}^{\rm L} + {v^{\ast}_{ki}}^{\rm U})\leq 1 
    \end{align}
$$

(32)式より
$$
    \begin{align}
        ~~ & ~~ \sum_{i\in N} \frac{w_{ki}^{\rm L} + w_{ki}^{\rm U}}{2} = \frac{1}{p}\sum_{i\in N} \frac{{v^{\ast}_{ki}}^{\rm L} + {v^{\ast}_{ki}}^{\rm U}}{2} = 1
    \end{align}
$$

よって $w_{ki}^{\rm L}, w_{ki}^{\rm U}$ は問題(10)の全ての制約を満たす。  
ここで(39)式を両辺 $\frac{1}{p}$ 倍することで
$$
    \begin{align}
        ~~ & ~~ \sum_{i\in N} ({w^{\ast}_{ki}}^{\rm L} - {w^{\ast}_{ki}}^{\rm U}) > \sum_{i\in N} (w_{ki}^{\rm L} - w_{ki}^{\rm U})
    \end{align}
$$
問題(10)において、$w_{k}$ の目的関数値が $w_{k}^{\ast}$ の目的関数値より小さくなるが、これは元々 $w_{k}^{\ast}$ が問題(10)の最適解であることに矛盾する。  
よって、問題(10)の最適解が問題(9)において最適ではないという仮定は矛盾しており、問題(10)の最適解は問題(9)においても最適解であるという結論が導かれる。  
また、問題(9)の最適解が問題(10)の最適解になることも同様にして導くことができる。  
以上より、問題(9)と問題(10)は同値な問題である。
