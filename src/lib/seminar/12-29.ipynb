### 問１
以下のよう問いがあったとき、期待値を求め、期待効用にもとづいた判断基準を示せ。「確率0.1で1000円当たり、確率0.4で500円当たり、確率0.5で100円当たるクジを考えます。このクジの値段が300円のとき、あなたはクジを買いますか?それはなぜですか?」

わかりました、問題1から解説を交えながら解答を進めましょう。

クジには以下の賞金があります：
- 確率 0.1 で 1000 円
- 確率 0.4 で 500 円
- 確率 0.5 で 100 円

クジの期待値を計算するには、それぞれの賞金にその確率を掛けて、それらを合計します。

期待値 $E$ は次のようになります。

$E = (0.1 \times 1000) + (0.4 \times 500) + (0.5 \times 100)  $

期待効用理論では、賞金の金額だけでなく、その賞金を得ることに対する個人の主観的価値も考慮に入れます。効用関数はこの主観的価値を数値化するもので、通常は金額の関数として表されます。ただし、この問題では効用関数は与えられていません。多くの場合、効用は対数的またはルート的な形で増加するとされますが、ここでは単純に金額の期待値で判断します。

問題文に「クジの値段が300円」とありますので、この価値（期待値）がクジの値段を上回れば、クジを買うことは理論的には価値があると判断できます。

それでは、具体的な計算を進めましょう。

期待値 $E$ は350円です。クジの値段が300円なので、期待値はクジの値段を50円上回ります。これは、長期的に見れば、このクジを買うことで平均して1回あたり50円の利益が得られることを意味します。したがって、純粋に期待値を基準に判断する場合、クジを買うのは理論的に価値があると言えます。

### 問２

ある個人の毎年の所得は不確実で、25%の確率で400万、75%の確率で1600万になる。この個人の効用関数が$ U(x) = \sqrt{x} $、で与えられている。  
(1)個人の一年の期待所得と期待効用を求めよ。  
(2)この不確実な状況に対して、毎年確実なy円の所得が得られる状況を比較した場合。この個人が毎年のy円を好むためには、yはどの範囲になくてはいけないか？  


問題2では、個人の毎年の所得が不確実で、25%の確率で400万円、75%の確率で1600万円になるとのことです。効用関数は $U(x) = \sqrt{x} $で与えられています。この情報を基に、次の2つの値を求めます。

1. 個人の一年の期待所得
2. 個人の一年の期待効用

そして、確実な年収 $y$ に対して、その人がどの範囲の$y$であれば、不確実な所得よりも確実な所得を好むかを判断します。

#### 期待所得の計算:

期待所得 $E$ は、各所得とその発生確率の積の合計で計算されます。

$E = (0.25 \times 400万) + (0.75 \times 1600万)$

期待所得 $E$ は1,300万円です。

#### 期待効用の計算:

期待効用 $EU$ は、各所得の効用とその発生確率の積の合計で計算されます。

$EU = (0.25 \times \sqrt{400万}) + (0.75 \times \sqrt{1600万})$

期待効用 $EU$ は3,500です（単位は万円の平方根）。

#### 確実な年収 $ y $ に対する期待効用の比較:

確実な年収 $y$ の効用は $ U(y) = \sqrt{y} $ です。この人が不確実な所得よりも確実な所得 $y$ を好むためには、$ \sqrt{y} $が $EU$ よりも大きくなくてはなりません。


次に、この期待効用に等しい確実な所得 $y$ を見つける必要があります。つまり、次の方程式を満たす $y$ を求めることになります。

$\sqrt{y} = EU$

この方程式を解いて $y$ を求めます。また、この個人が不確実な所得よりも確実な所得を好むためには、$y$ は求めた値以上でなければなりません。それでは、$y$を計算してみましょう。

求めた確実な所得 $y$ は1,225万円です。

これは、不確実な状況下での期待効用と同じ効用をもたらす確実な所得の額です。したがって、この個人が不確実な所得(25%の確率で400万円、75%の確率で1600万円）よりも確実な所得を好むためには、その所得は少なくとも1,225万円でなくてはなりません。これは、リスクに対する態度がリスク中立であると仮定した場合の結果です。

つまり求める範囲は
$y\geq 1225万 $

### 問３

ある個人は職業A と職業Bの選択に迫られている。Aは0.5の確率で、900万か確率0.5で1600万の賃金$×$が得られ、職業Bは確実な賃金$y$が得られる。この個人は効用関数$u(×)= \sqrt{x}$ 、とする。このとき、$y$がいくら以上ならば職業Bを選択するか?

問題3では、個人が職業Aと職業Bの選択を迫られており、職業Aは0.5の確率で900万円、または0.5の確率で1600万円の賃金を得ることができ、職業Bは確実に賃金$y$を得られるとのことです。効用関数は$ U(x)=\sqrt{x}$と与えられています。

この情報を基に、次の値を求めます。
1. 職業Aにおける期待効用
2. 職業Bにおける効用

そして、個人が職業Bを選択するための賃金 $y$の最低額を求めます。

#### 職業Aの期待効用の計算:

職業Aの期待効用$EU_A$は次のように計算されます。

$EU_A = (0.5 \times \sqrt{900万}) + (0.5 \times \sqrt{1600万})$

職業Aの期待効用 \( EU_A \) は3,500です（単位は万円の平方根）。

#### 職業Bの効用の計算:

職業Bの効用$U_B$は次のように計算されます。

$U_B = \sqrt{y}$

#### 職業Bを選択する条件:

個人が職業Bを選択するためには、$U_B \geq EU_A$となる最小の$y$を見つける必要があります。

次に、職業Bを選択するための賃金$y$の最低額を求めます。これは、個人が職業Bの確実な効用$U_B = \sqrt{y}$が職業Aの期待効用$EU_A$と少なくとも等しくなる$y$の値です。つまり、次の方程式を解くことになります。

$\sqrt{y} = 3500$

この方程式を解いて$y$を求めます。

職業Bを選択するための賃金$y$の最低額は1,225万円です。

これは、職業Aの不確実な収入に対する期待効用と同じ効用をもたらす、職業Bの確実な賃金の額です。したがって、この個人が職業Bを選択するためには、その賃金は少なくとも1,225万円でなくてはなりません。

つまり求める範囲は
$y\geq 1225万 $

### 問４

富の額を$W$として、ある消費者の効用関数が、$U(W)=W^{0.5}$ (累乗をで記述メ^2でXの2乗)で示されるとする。この消費者の現在の富の保有額は10,000であり、事故が発生する可能性を考慮して、事前に保険料を支払うと、事故が起きたときに損害を全額保障するような保険を考える。事故が起きる確率を5%、事故が起きたときの損害を10,000とした場合、この消費者が保険に加入するのは、保険料がいくら以下のときか。ただし、この消費者は期待効用の最大化を図るものとし、また自分が事故に会う確率を知っているものとする。なお、事故が起きなくても保険料は返却されない。

問題4では、消費者の効用関数が $U(W) = W^{0.5}$で与えられており、消費者の現在の富の保有額は10,000です。事故が発生する確率は5%で、事故が起きたときの損害は10,000となります。保険に加入するかどうかの判断を、期待効用の最大化の観点から考えます。

#### 期待効用の計算：

保険に加入しない場合の期待効用#EU_{no-insurance}$は、事故が起きない場合と起きた場合の効用の期待値です。

$EU_{no-insurance} = 0.95 \times U(10000) + 0.05 \times U(0)$

保険に加入する場合、消費者は損害を全額補償する保険を考えます。保険料を$P$としたとき、期待効用$EU_{insurance}$は以下の通りです。

$EU_{insurance} = U(10000 - P)$

#### 保険に加入する条件：

消費者が保険に加入するためには、加入しない場合の期待効用よりも加入した場合の期待効用の方が大きくなくてはなりません。

$EU_{insurance} \geq EU_{no-insurance}$

保険料 $P$がいくら以下であれば、消費者は保険に加入するかを求めるために、上記の不等式を$P$について解きます。

それでは、まず保険に加入しない場合の期待効用を計算し、次に保険に加入する条件を求めます。

保険に加入しない場合の期待効用$EU_{no-insurance}$は95です。

次に、保険料$P$がいくら以下であれば消費者は保険に加入するかを求めます。これは、次の不等式を満たす$P$を見つけることに相当します。

$U(10000 - P) \geq 95$

$(10000 - P)^{0.5} \geq 95$

この不等式を解いて、保険料$P$の最大値を計算します。

保険料$P$が975以下であれば、消費者は保険に加入することが期待効用を最大化する選択です。

これは、保険料が975を超えると、保険に加入した場合の効用が保険に加入しない場合の期待効用95を下回るため、保険に加入するインセンティブがなくなることを意味します。保険料が975以下であれば、消費者は事故が発生した場合のリスクを回避できるため、保険に加入することが望ましいと判断されます。

### 問５

以下の問にプロスペクト理論の考えを用いて解答してください。
質問1:あなたの目の前に、以下の二つの選択肢が提示されたものとする。  
選択肢A:1万円が無条件で手に入る。  
選択肢B:コイントスをし、表が出たら2万円が手に入るが、裏が出たら何も手に入らない。  
質問2:あなたは2万円の借金を抱えているものとする。そのとき、同様に以下の二つの選択肢が提示されたものとする。  
選択肢A:無条件で負債が1万円減額され、借金の総額が1万円となる。  
選択肢B:コインを投げ、表が出たら借金の支払い2万円が全額免除されるが、裏が出たら借金の総額は変わらない  
(1).価值関数$V(X)=PX^{0.5} , if X\geq 0, -0.5(-PX^{0.5}), if X<0$という損失回避の価値関数を持つ個人は上の質問1質問2のそれぞれの選択肢でどのよう価値の大きになるかを計算で示せ。(確率は期待効用のようにそのままかけて良い)  
(2)(1)の結果にもとづいて、それぞれの質問ではどちらの選択肢を選択するかを示せ。  

問題5ではプロスペクト理論を用いて解答します。プロスペクト理論は、人々がリスクに直面したときの意思決定に関する理論で、特に損失回避の傾向をモデル化するために使用されます。

価値関数は次のように定義されています：
$V(X) = PX^{0.5} \text{ if } X \geq 0, \text{ and } -0.5(-PX^{0.5}) \text{ if } X < 0$
ここで$P$は確率を表します。

#### 質問1の解析:

選択肢A：1万円が無条件で手に入る。
選択肢B：コイントスをし、表が出たら2万円が手に入るが、裏が出たら何も手に入らない。

##### 選択肢Aの価値:

選択肢Aは確実なので、価値は$1 \times 10000^{0.5}$となります。

##### 選択肢Bの価値:

選択肢Bは確率$P = 0.5$で2万円を得るため、価値は$0.5 \times 20000^{0.5}$となります。

#### 質問2の解析:

選択肢A：無条件で負債が1万円減額される。
選択肢B：コインを投げ、表が出たら借金の支払い2万円が全額免除されるが、裏が出たら借金の総額は変わらない。

##### 選択肢Aの価値:

選択肢Aは確実な減額なので、価値は$-0.5 \times (-1 \times 10000^{0.5})$となります。

##### 選択肢Bの価値:

選択肢Bは確率$P = 0.5$で借金が全額免除されるため、価値は$V(B) = 0.5 \times -0.5 \times (-20000^{0.5})$となります。

質問1と質問2の選択肢の価値を再計算した結果は以下の通りです。

#### 質問1:

- 選択肢Aの価値: 100
- 選択肢Bの価値: 約70.71

#### 質問2:

- 選択肢Aの価値: 50
- 選択肢Bの価値: 35.36

#### 解説:
質問1では、選択肢A（1万円が確実に手に入る）が選択肢B（50%の確率で2万円が手に入る）よりも価値が高いとされます。これは、プロスペクト理論における確実性効果を示しており、人々は確実な小さな利得を大きな不確実な利得よりも好む傾向があります。

質問2では、選択肢A（負債が1万円減額される）の価値は50、選択肢B（50%の確率で2万円の借金が全額免除される）の価値は約35.36です。この結果は、プロスペクト理論における損失回避性と確実性効果を示しています。

これらの結果から、質問1では選択肢A、質問2では選択肢Aを選ぶことが、プロスペクト理論に基づく合理的な選択となります。これは、得られるものに対しては確実性を好み、失うものに対してはリスクを避ける傾向があることを反映しています。

### 問６

合理的期待にもとづいて参照点を考えて意思決定をしている例をスポーツのシーンで挙げよ。

合理的期待理論は、人々が将来のアウトカムに対して形成する期待が、その人の意思決定にどのように影響を与えるかを説明します。

スポーツにおける典型的な例は、選手が次の試合やシーズンのパフォーマンスに対する期待をどのように設定するかです。選手やチームは過去の成績、トレーニングの強度、対戦相手の強さなどを基にして、次の試合やシーズンにおける自分たちのパフォーマンスに対する合理的な期待を形成します。そして、その期待は彼らのトレーニングの方法、戦略の立案、メンタルの準備に影響を及ぼします。

例えば、サッカーチームがリーグの強豪チームと対戦する前に、自チームが勝つためには少なくとも3点を取る必要があると考えたとします。この得点期待は、彼らが攻撃的な戦術を採用するか、守備に重点を置くかといった戦略的な意思決定に影響を与えるでしょう。合理的な期待は、現実的な目標設定に基づいており、過去のデータや統計に裏打ちされています。

このように、スポーツ選手やチームが合理的期待を設定し、それに基づいて戦略を立て、準備を行うことは、成功に向けた意思決定の重要な要素です。
