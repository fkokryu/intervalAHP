### それぞれの手法の特性・意味合い

・perfect incoprporation

どんな時でも合意の区間を持つ

全員の区間重要度を含む広い区間重要度→誰も同意しない部分を含んでいる可能性もある

・common ground

DM全員が共通部分を持つ時だけ合意の区間を持つ

合意の区間内のどの重要度の組み合わせでも全員が同意する

・partial incorporation

どんな時でも合意の区間を持つ

区間内のある重要度に対して少なくとも1人は同意する
