クジの価格が300円であるため、期待値は350円となり、クジの価格を50円上回る。従って、期待値を基準に判断する場合、クジを購入することは理論的に価値があると判断される。これは、長期的に見れば、このクジを購入することで平均して1回あたり50円の利益が得られることを意味する。

#### 問２

ある個人の毎年の所得は不確実で、$25\%$の確率で$400$万円、$75\%$の確率で$1600$万円になる。この個人の効用関数は $U(x) = \sqrt{x}$ で与えられている。

(1) 個人の一年の期待所得と期待効用を求める。

- **期待所得**
  $$
  E = (0.25 \times 400万) + (0.75 \times 1600万) = 1300万円
  $$

- **期待効用**
  $$
  EU = (0.25 \times \sqrt{400万}) + (0.75 \times \sqrt{1600万})
  $$
  この計算結果は約3500万円の平方根である。

(2) この不確実な状況に対して、毎年確実な$y$円の所得が得られる状況を比較する。

- 確実な年収$y$の効用は $U(y) = \sqrt{y}$ である。この人が不確実な所得よりも確実な所得$y$を好むためには、$\sqrt{y}$が$EU$よりも大きくなくてはならない。
  $$
  \sqrt{y} \geq EU
  $$
  $$
  y \geq (EU)^2
  $$

- $EU$ は約3500万円の平方根なので、$y$は $(3500万円の平方根)^2$ 以上でなければならない。この計算結果は、$1225万円$以上である。

したがって、この個人が不確実な所得よりも確実な所得を好むためには、その所得は少なくとも$1225万円$でなくてはならない。これはリスクに対する態度がリスク中立であると仮定した場合の結果である。

求める範囲は $y \geq 1225万円$ である。

#### 問３

個人が職業Aと職業Bの選択に迫られている状況について考える。職業Aでは$0.5$の確率で$900$万円、または$0.5$の確率で$1600$万円の賃金を得ることができ、職業Bでは確実に賃金$y$を得られる。効用関数は$U(x) = \sqrt{x}$とする。

1. **職業Aの期待効用の計算**
   $$
   EU_A = (0.5 \times \sqrt{900万}) + (0.5 \times \sqrt{1600万})
   $$

2. **職業Bの効用の計算**
   $$
   U_B = \sqrt{y}
   $$

3. **職業Bを選択する条件**
   個人が職業Bを選択するためには、$U_B \geq EU_A$となる最小の$y$を見つける必要がある。これは、職業Bの確実な効用$U_B = \sqrt{y}$が職業Aの期待効用$EU_A$と少なくとも等しくなる$y$の値を求めることを意味する。

   $$
   \sqrt{y} \geq EU_A
   $$
   $$
   y \geq (EU_A)^2
   $$

   この方程式を解いて$y$を求める。

職業Bを選択するための賃金$y$の最低額は$1225$万円である。これは、職業Aの不確実な収入に対する期待効用と同じ効用をもたらす、職業Bの確実な賃金の額である。したがって、この個人が職業Bを選択するためには、その賃金は少なくとも$1225$万円でなければならない。

したがって、求める範囲は $y \geq 1225万円$ である。

#### 問４

消費者の効用関数が $U(W) = W^{0.5}$ であり、現在の富の保有額が $10,000$、事故が発生する確率が $5\%$、事故が起きたときの損害が $10,000$ とする。事前に保険料を支払って、事故が起きたときに損害を全額保障する保険について考える。保険料を $P$ とした場合、消費者が保険に加入するのは、保険料がいくら以下のときかを求める。

1. **保険に加入しない場合の期待効用**
   $$
   EU_{\text{no-insurance}} = 0.95 \times U(10000) + 0.05 \times U(0) = 0.95 \times 100^{0.5} + 0.05 \times 0
   $$

2. **保険に加入する場合の期待効用**
   $$
   EU_{\text{insurance}} = U(10000 - P)
   $$

3. **保険に加入する条件**
   消費者が保険に加入するためには、加入しない場合の期待効用よりも加入した場合の期待効用の方が大きくなくてはならない。
   $$
   EU_{\text{insurance}} \geq EU_{\text{no-insurance}}
   $$
   $$
   U(10000 - P) \geq EU_{\text{no-insurance}}
   $$
   $$
   (10000 - P)^{0.5} \geq EU_{\text{no-insurance}}
   $$

   この不等式を $P$ について解くと、保険料がいくら以下であれば消費者は保険に加入するかが分かる。

保険に加入しない場合の期待効用 $EU_{\text{no-insurance}}$ は約 $95$ である。

次に、保険料 $P$ がいくら以下であれば消費者は保険に加入するかを求める。これは次の不等式を満たす $P$ を見つけることに相当する。
$$
(10000 - P)^{0.5} \geq 95
$$

この不等式を解いて、保険料 $P$ の最大値を求める。

保険料 $P$ が $975$ 以下であれば、消費者は保険に加入することが期待効用を最大化する選択である。これは、保険料が $975$ を超えると、保険に加入した場合の効用が保険に加入しない場合の期待効用 $95$ を下回るため、保険に加入するインセンティブがなくなることを意味する。保険料が $975$ 以下であれば、消費者は事故が発生した場合のリスクを回避できるため、保険に加入することが望ましいと判断される。

#### 問５

プロスペクト理論を用いて解答する。価値関数は以下のように定義される：
$V(X) = PX^{0.5} \text{ if } X \geq 0, \text{ and } -0.5(-PX^{0.5}) \text{ if } X < 0$

##### 質問1の解析:

選択肢A: 1万円が無条件で手に入る。  
選択肢B: コイントスをし、表が出たら2万円が手に入るが、裏が出たら何も手に入らない。

###### 選択肢Aの価値:
$$
V(A) = 1 \times 10000^{0.5} = 100
$$

###### 選択肢Bの価値:
$$
V(B) = 0.5 \times 20000^{0.5} = 70.71
$$

##### 質問2の解析:

選択肢A: 無条件で負債が1万円減額される。  
選択肢B: コインを投げ、表が出たら借金の支払い2万円が全額免除されるが、裏が出たら借金の総額は変わらない。

###### 選択肢Aの価値:
$$
V(A) = -0.5 \times (-1 \times 10000^{0.5}) = 50
$$

###### 選択肢Bの価値:
$$
V(B) = -0.5 \times -0.5 \times (-20000^{0.5}) = 35.36
$$

##### (2) どちらの選択肢を選択するか:

- **質問1**: 選択肢Aの価値（$100$）が選択肢Bの価値（$70.71$）よりも高いため、選択肢Aを選択することが期待される。
- **質問2**: 選択肢Aの価値（$50$）が選択肢Bの価値（$35.36$）よりも高いため、ここでも選択肢Aを選択することが期待される。

これらの結果から、プロスペクト理論に基づくと、質問1では確実な小さな利得を、質問2では確実な小さな損失減少を選択することが合理的である。これは、得るものに対してリスク回避的、失うものに対してリスクを避ける傾向を示している。

###### ナッシュ均衡:
- (A, A): 二郎も太郎も戦略Aを選ぶとき、互いに戦略を変更することで利得が減少するため、これがナッシュ均衡である。

###### 支配戦略均衡:
- (A, A): 二郎にとってAはBを支配し、太郎にとってもAはBを支配するため、(A, A)が支配戦略均衡である。

##### 右のゲームの支払い行列:
```
| 花子 \ たけし | A     | B     |
|--------------|-------|-------|
| A            | (3, 1)| (0, 0)|
| B            | (0, 0)| (1, 3)|
```

###### ナッシュ均衡:
- (A, B): 花子がAを選び、たけしがBを選ぶとき、互いに戦略を変更することで利得が減少するため、これがナッシュ均衡である。

- 存在しない: このゲームには、一方の戦略が他方を常に支配するような戦略が存在しないため、支配戦略均衡は存在しない。

#### 問２

同質財を生産するA社とB社の市場を考え、逆需要関数と費用関数を用いて以下の問いに答える。

##### 逆需要関数と費用関数：
- 逆需要関数: $P(X) = 100 - X$
- 費用関数: $C(X) = 10X$
- 限界費用: $MC(X) = 10$

##### 1. 均衡市場価格の導出：
均衡市場価格は、市場の需要と供給が等しくなる点で決まる。ここで、限界費用が一定（10）なので、供給曲線は価格10で水平になる。したがって、均衡市場価格は10円となる。

##### 2. P=50のとき均衡にならないことの証明：
市場価格が50円の場合、需要量は$100 - P = 100 - 50 = 50$となる。しかし、この価格では限界費用（10円）を大きく上回るため、企業はより多くの量を供給しようとする。結果として、供給量の増加により市場価格は下落し、市場は均衡に達しない。

##### 3. P=5でも均衡にならないことの証明：
市場価格が5円の場合、限界費用（10円）を下回るため、企業は損失を出してまで供給するインセンティブがない。結果として、供給量はゼロになり、市場は均衡に達しない。

##### 4. ベルトラン競争の例とモデルの意図の説明：
ベルトラン競争は価格競争を特徴とする。このモデルでは、企業は価格を限界費用まで下げる傾向がある。長期的には企業の利益はゼロになる可能性があり、これは消費者には低価格での利益をもたらすが、企業の利益は圧迫される。このモデルの目的は、価格の下落圧力と企業間の競争の激しさを示すことにあり、市場の効率性を高める一方で、企業の利益を犠牲にすることを示している。

添付の図(問２.png)では、逆需要関数 $P(X) = 100 - X$ に基づく需要曲線と、一定の限界費用 $MC = 10$ に基づく水平な供給曲線が示されている。均衡点では、需要と供給が一致し、市場価格は10円となる。また、市場価格が50円や5円の場合、市場が均衡に達しない様子が示されている。
