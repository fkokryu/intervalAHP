### 問１
スライド6のP27にある2つのゲームのナッシュ均衡と支配戦略均衡を求めよ。

問題1に基づいて、ゲームのナッシュ均衡と支配戦略均衡を求めます。

与えられたゲームは次の支払い行列を持っています：

左のゲームの支払い行列:
```
| 二郎 \ 太郎 | A     | B     |
|------------|-------|-------|
| A          | (4, 4)| (0, 1)|
| B          | (1, 0)| (1, 1)|
```

右のゲームの支払い行列:
```
| 花子 \ たけし | A     | B     |
|--------------|-------|-------|
| A            | (3, 1)| (0, 0)|
| B            | (0, 0)| (1, 3)|
```

### 解答と解説:

#### 左のゲーム:
- **ナッシュ均衡**: 二郎と太郎がともに戦略Aを選択するとき、両プレイヤーともに戦略を変えるインセンティブがありません。なぜなら、どちらのプレイヤーも戦略を変更しても利得が減少するからです。したがって、(A, A)はナッシュ均衡です。
- **支配戦略均衡**: 二郎にとって戦略AはBを支配しています（4 > 1）、太郎にとっても戦略AはBを支配しています（4 > 1）。したがって、(A, A)は支配戦略均衡です。

#### 右のゲーム:
- **ナッシュ均衡**: このゲームには二つのナッシュ均衡が存在します。花子がAを選び、たけしがBを選ぶ場合（(A, B)）、または花子がBを選び、たけしがAを選ぶ場合（(B, A)）です。どちらの場合も、一方のプレイヤーが戦略を変更すると、他方も戦略を変更するインセンティブが生じます。
- **支配戦略均衡**: このゲームには支配戦略が存在しません。どの戦略も明確に他の戦略より優れているわけではないため、支配戦略均衡はありません。

つまり、

左のゲームのナッシュ均衡と支配戦略均衡: $(A, A)$  
右のゲームのナッシュ均衡: $(A, B)$, $(B, A)$  
右のゲームの支配戦略均衡: なし

これらの結果は、プレイヤーが自分の利得を最大化することを目指すという基本的な仮定に基づいています。ナッシュ均衡では、各プレイヤーが相手の戦略を考慮に入れて自分の最適な戦略を選択します。一方、支配戦略均衡は、他のすべての戦略に対して常に優れた結果をもたらす戦略が存在する場合に定義されます。

### 問２

いま、同質財をAとBの2社が生産する市場を考える。  
逆需要関数:$P(X)100-X$  
費用関数:$C(X)10X \Rightarrow \text{限界費用：} MC(X) = 10$  
このとき図解をしながら以下の問いに答えよ。

1.均衡市場価格はいくらになるか。  
2.P50のとき均衡にならないことを示せ。  
3.P=5でも均衡にならないことを示せ。  
4.ベルトラン競争の例とこのモデルの意図を説明せよ。

問題2に対する解答です。市場には企業AとBが同質財を生産しており、逆需要関数は$P(X) = 100 - X $であり、費用関数は$C(X) = 10X$です。限界費用（MC）は10円となります。このときの均衡市場価格と各企業の戦略を考えます。

1. **均衡市場価格の導出**:
均衡市場価格は、企業が限界費用と同じ価格で供給するときに成立します。企業が利益を最大化するためには、価格は限界費用に等しくなるので、均衡価格$P^*$は10円になります。

2. **$P = 50$のとき均衡にならないことの証明**:
もし市場価格が50円であれば、この価格で供給される量は$100 - P = 100 - 50 = 50$になりますが、企業の限界費用は10円なので、市場価格が限界費用よりも高いと企業はより多くの量を供給しようとするため、この価格では市場は均衡に達しません。

3. **$P = 5$でも均衡にならないことの証明**:
市場価格が5円であれば、企業は費用を上回る価格で供給するインセンティブがなくなります。限界費用が10円なので、損失を出してまで供給する理由がないため、この価格では均衡にはなりません。

4. **ベルトラン競争の例とモデルの意図の説明**:
ベルトラン競争は価格競争のモデルで、企業は価格を武器にして市場シェアを奪い合います。このモデルでは、企業は価格を限界費用まで下げてしまうため、長期的には企業の利益はゼロになります。この結果は価格戦争の激しさを示しており、消費者には低価格での利益をもたらすものの、企業の利益は圧迫されることを意味しています。このモデルは価格の下落圧力と企業間の競争の激しさを示し、市場の効率性を高める一方で、企業の利益を犠牲にすることを示しています。

#### 図の説明
上記のグラフは、問題2に関連する市場の図解を示しています。

- **需要曲線 (P = 100 - X)**: 価格$P$と数量$X$の関係を示す曲線。価格が高いと需要の数量が減少します。

- **供給曲線 (P = 10)**: この場合、供給曲線は限界費用が10円で一定であるため、価格$P = 10$で水平です。

- **均衡点**: 需要曲線と供給曲線が交わる点。この場合、均衡数量は90、均衡価格は10円です。

- **P = 50の場合**: 赤い点線。この価格では市場は均衡に達しません。供給はこの価格で増加しますが、需要は減少します。

- **P = 5の場合**: 緑の点線。この価格では市場は均衡に達しません。限界費用を下回るため、供給はゼロになります。

### 問３

中位投票者定理.いま、0:左翼(リベラル)、1:右翼(保守)として、2大政党制で、議席数を争う選挙の主張を考える。世論は(0,1)区間上に一様分布している。有権者は自分の考えに近い方に投票するものとする。  
1.このゲームを定義せよ。  
2.均衡が1/2になることを証明せよ。

問題3に対する解答と解説を以下に記述します。

### 問題3: 中位投票者定理

#### ゲームの定義:

- **プレイヤー**: 二つの政党、左翼（リベラル：0）と右翼（保守：1）。
- **戦略**: それぞれの政党は、有権者が一様分布している$[0,1]$区間上で政策を選択します。
- **ペイオフ**: 最も多くの票を獲得することで、議席数を最大化する。

#### 均衡が1/2になることの証明:

- 有権者は自分の意見に最も近い政党に投票するため、中間点（0.5）に最も近い位置に政策を設定する政党が最も多くの票を獲得します。
- 有権者の意見が$[0,1]$区間で一様分布しているため、中位の有権者は0.5の位置にいます。そのため、両政党は中位の有権者を引き込むために、自身の政策を0.5の位置に設定するインセンティブがあります。
- 一方の政党が0.5から逸脱すると、他方の政党は中位の有権者により近づくことで、より多くの票を獲得できます。
- したがって、ナッシュ均衡は両政党が0.5の位置に政策を設定することです。

これにより、選挙戦のゲームにおいて中位投票者定理が成り立ち、均衡点が0.5になることが示されます。この理論は政党が選挙で勝利するためには中立的な立場を取ることが最適であることを示しており、実際の選挙戦でも多くの場合、政党は中道的な政策を採用する傾向にあります。

### 問題4: エスカレーターにおける立ち止まる場所の選択

1. **プレイヤーAの行動**: Aは100円単位で0円から1000円まで、Bに提案する金額を選びます。これにより、10の可能な行動（0, 100, 200, ..., 900, 1000円）があります。

1. **Bの最適な戦略**: Bにとって、受け取る金額が0円より多ければ、提案を受け入れることが最適です。そのため、BはAからのいかなる正の提案も受け入れると想定されます。ただし、提案が0円の場合、Bは受け入れても拒否しても同じ結果（0円）となるため、無差別ですが、拒否すると想定します。

部分ゲーム完全均衡では、プレイヤーAは100円を提案し、プレイヤーBはそれを受け入れます。この結果、Aは900円、Bは100円を受け取ります。

### 問６

最後通牒ゲームに不平等回避モデルのプレイヤーを導入してみよう。  
①プレイヤーAがプレイヤーBに対して、100円単位で1000円の中から提案する  
②プレイヤーBはAの提案額を「受入れ」か「拒否」の選択をする。  
-受入れの場合「Aの提案額が分配される」  
-拒否の場合「お互いに0円、0円の分配となる」  
スライドにある設定の不平等回避モデル$Ui(x)=xi-\alpha (|xj-xi|)-\beta (|xi-xj|)$  
($\alpha \ 0.8, \beta \ 0.3$ ij=A,Bを想定)を考えて、金銭的利得ではなく、心理的利得を1000:0から順に5:5や0:1000まで計算し、心理的利得による部分ゲーム完全均衡をもとめよう。

この問題は、最後通牒ゲームに不平等回避モデルを導入して、心理的利得に基づいた部分ゲーム完全均衡を求めるものです。不平等回避モデルでは、プレイヤーの利得は単に金銭的な利得ではなく、他者との比較に基づく心理的な影響も含まれます。

### 不平等回避モデルの定義

モデルは次のように定義されます：

$ U_i(x) = x_i - \alpha (|x_j - x_i|) - \beta (|x_i - x_j|)$

ここで、$x_i$ はプレイヤー$i$の利得$\alpha$ と$\beta$は不平等回避に関するパラメータで、それぞれが他者より多く、または少なく得る場合の心理的損失を表します。この問題では$\alpha = 0.8$、$\beta = 0.3$とされています。

### 心理的利得の計算

心理的利得は、異なる提案に対するプレイヤーAとBの利得を計算することで求めます。例えば、プレイヤーAがプレイヤーBに100円を提案し、残りの900円を自分が得る場合、心理的利得は次のように計算されます：

- プレイヤーAの心理的利得: $U_A(900) = 900 - 0.8 * |900 - 100| - 0.3 * |900 - 100|$
- プレイヤーBの心理的利得: $U_B(100) = 100 - 0.8 * |100 - 900| - 0.3 * |100 - 900|$

この計算を、異なる提案（100円単位で0円から1000円まで）に対して行い、それぞれの提案に対する心理的利得を求めます。

プレイヤーAの提案とそれに対するプレイヤーBの心理的利得を計算し、Bが提案を受け入れるか拒否するかを判断します。その後、部分ゲーム完全均衡を見つけます。

心理的利得の計算結果は以下の通りです：

プレイヤーAの提案とそれに対する両プレイヤーの心理的利得：
- 提案が0円の場合、Aの利得は -100、Bの利得は -1100。
- 提案が100円の場合、Aの利得は 20、Bの利得は -780。
- 提案が200円の場合、Aの利得は 140、Bの利得は -460。
- ...
- 提案が900円の場合、Aの利得は -780、Bの利得は 20。
- 提案が1000円の場合、Aの利得は -1100、Bの利得は -100。

この結果から、提案が500円の場合、両プレイヤーの心理的利得が500となり、これが最大です。したがって、部分ゲーム完全均衡では、プレイヤーAはプレイヤーBに500円を提案し、プレイヤーBはそれを受け入れることが予想されます。この提案は、両者にとって心理的利得が最大になるため、最適な戦略となります。

### 問７

1000円を100円単位で105ペアによる独裁者ゲームをする状況を考える。社会規範モデルを考える。  
周りの独裁者役が4人全員5:5にしているとき、iさんが8:2にするときと、5:5にするときを比較せよ。$\gamma \ 0.8$を仮定。

この問題では、105ペアによる独裁者ゲームを考え、社会規範モデルを用いて、iさんが異なる分配を選択した場合の結果を比較します。社会規範モデルでは、個人の利得は単に金銭的な利得だけでなく、社会規範との一致による心理的な要素も含まれます。この場合、社会規範として「5:5の分配」が設定されており、逸脱すると心理的なコストが発生すると仮定します。パラメータ$\gamma$は0.8とされています。

#### 問題の解析

1. **iさんが5:5に分配する場合**:
   - 金銭的利得は500円。
   - 社会規範に従っているため、心理的なコストは発生しない。
   - 合計利得は500円。

2. **iさんが8:2に分配する場合**:
   - 金銭的利得は800円。
   - 社会規範からの逸脱による心理的なコストを考慮。このコストは逸脱の度合いに比例し、$\gamma$によって重み付けされる。逸脱の度合いは、標準的な分配（500円）との差額（300円）となる。
   - 心理的コストは$\gamma \times 300 = 0.8 \times 300 = 240$円。
   - 合計利得は$800 - 240 = 560$円。

#### 結論

iさんが5:5に分配する場合、合計利得は500円です。一方、8:2に分配する場合、合計利得は560円となります。この結果から、金銭的利得を最大化することを最優先する場合、iさんは8:2の分配を選択することが合理的です。

### 問８

同調行動を起こす環境についてサッカーの観衆とレフェリー以外にどのようものがあるか。(労働環境、委員会活動、教育現場などなどで実例を挙げよ。)

同調行動は、個人が周囲の人々の行動や意見に合わせて自分の行動や意見を変える社会心理学の現象です。サッカーの観衆やレフェリー以外にも、様々な環境で同調行動が観察されます。以下にいくつかの実例を挙げます：

1. **労働環境**: 職場では、社員が他の同僚や上司の行動や意見に合わせることがあります。例えば、あるチームメンバーが遅くまで残業していると、他のメンバーも同様に残業をする傾向があります。また、上司の意見に対する同意が、他のメンバーの意見を形成する場合もあります。

2. **委員会活動**: 委員会やボードミーティングなどで、意思決定の過程において、少数意見の人々が多数派の意見に同調することがあります。これは「グループシンク」としても知られており、多数派の意見に異を唱えることが難しい環境が生まれることがあります。

3. **教育現場**: 学校や大学などの教育環境では、学生がクラスメートの行動や意見に影響を受けることがよくあります。たとえば、ある学生が特定のスタイルの服を着ると、他の学生もそのスタイルを真似することがあります。また、ある意見や回答がクラスで人気を博すと、他の学生もその意見に同調することがあります。

4. **社会的運動**: 社会的な傾向や運動、例えば環境保護や健康に関する運動などでも、人々は周囲の行動に影響を受け、同様の行動を取ることがあります。例えば、周囲の人々がリサイクルに積極的であると、他の人々もリサイクルを始めることがあります。

これらの例は、同調行動が様々な状況で発生し、個人の行動や意見形成に大きな影響を与えることを示しています。同調行動はグループ内の一体感を強化する一方で、個々の意見の多様性を低下させる可能性もあるため、バランスを取ることが重要です。

### 問９

コロナの状況においてマスクやククチン接種が社会的シグナリングであるという研究がある。囚人のジレンマゲームのような、2人のゲームを考えて、(マスク、マスク)は平常時だと均衡にならない(非コロナ時)が、社会的シグナリングのパラメータを考えて、コロナ時には(マスク、マスク)が均衡になるように、利得とパラメータを設定したゲームを作成して、ナッシュ均衡と社会的シグナリング均衡を説明せよ。

シミュレーションの結果、以下のようになりました：

- $\gamma$の値が 0.3 以下の場合、ナッシュ均衡は一貫して ("no_mask", "no_mask") となります。これは、どちらのプレイヤーもマスクを着用しないという結果で、社会的シグナリングが低い$\gamma$が低い）場合、マスクを着用することは均衡戦略ではありません。
- $\gamma$の値が 0.3 を超えると、社会的シグナリング均衡は ("mask", "mask") に移行し、両プレイヤーがマスクを着用する選択をします。これは、社会的シグナリングのパラメータ $\gammaが十分に高く、プレイヤーの戦略がパンデミック中にマスクを着用する方向に影響される状況を反映しています。

この結果は、社会的シグナリングが重要なゲームにおいて$\gamma$の値が均衡戦略にどのように影響を与えるかを示しています。与えられた利得構造では$\gamma$が増加するにつれて、マスクを着用する社会的インセンティブが個々のインセンティブを上回るほど強くなり、両プレイヤーの均衡が "no_mask" から "mask" へと変化することが明らかです。

シミュレーションの結果を基に、社会的シグナリングのパラメータ$\gamma$を0.5に設定します。この値は、社会的シグナリングの影響が強くなり、マスクを着用することが社会的に望ましい行動と認識されるコロナ禍の状況を表すため選択されました。また、他プレイヤーの選択が自分の効用に与える影響を示す$\alpha$を0.1と設定します。$\alpha$が小さい場合、他人の選択は自分の効用にそれほど大きな影響を与えません。

**利得表**:

$$
\begin{array}{c|c|c}
\text{プレイヤーA \ B} & \text{マスクを着用} & \text{マスクを着用しない} \\
\hline
\text{マスクを着用} & 3 \times 0.5, 3 \times 0.5 & 0, 5 \times 0.5 \\
\text{マスクを着用しない} & 5 \times 0.5, 0 & 0.5, 0.5 \\
\end{array}
$$

**ナッシュ均衡**:

- $\gamma = 0.5$の場合、マスクを着用する利得が他の戦略より高くなります。つまり、プレイヤーはマスクを着用することを選びます。均衡は ("mask", "mask") となります。

**社会的シグナリング均衡**:

- $\gamma$が0.5の場合、社会的シグナリングの効果が強くなります。これにより、プレイヤーはマスクを着用することで社会的な評価を高めようとします。したがって、社会的シグナリング均衡も ("mask", "mask") になります。

この解析から、$\gamma$の値が増加すると、プレイヤーの選択が社会的シグナリングの影響を受けやすくなることがわかります。コロナ禍のような特殊な状況では、社会的シグナリングが強く作用し、マスク着用という社会的に望ましい行動が均衡戦略となることを示しています。また、$\alpha$が低いことにより、他人の選択が自分の効用に与える影響が限定され、社会的シグナリングの効果がより顕著になる傾向があります。
